# üåç NDVI Time-Series Analysis

## What we'll do:
1. Fetch Sentinel-2 satellite data
2. Calculate NDVI (Vegetation Index)
3. Compare two time periods
4. Visualize the changes

**NDVI = (NIR - Red) / (NIR + Red)**

- High NDVI (0.6-1.0) ‚Üí Dense vegetation üå≤
- Low NDVI (0-0.3) ‚Üí Sparse vegetation or bare soil üèúÔ∏è

```python
import ee
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO

# Initialize Earth Engine
ee.Initialize(project='YOUR PROJECT ID') # YOUR PROJECT-ID

print("‚úÖ Earth Engine initialized!")
```

```python
# Poikilo Oros Forest, Chaidari, Attica, Greece
# You can change coordinates for other areas!

area = ee.Geometry.Rectangle([23.660, 38.030, 23.690, 38.055])

print("üìç Area of Interest: Poikilo Oros Forest, Petroupoli")
print(f"   Coordinates: {area.coordinates().getInfo()}")
```

```python
# Summer 2020
image_2020 = ee.ImageCollection('COPERNICUS/S2_SR') \
    .filterBounds(area) \
    .filterDate('2020-06-01', '2020-08-31') \
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20)) \
    .median()

# Summer 2024
image_2024 = ee.ImageCollection('COPERNICUS/S2_SR') \
    .filterBounds(area) \
    .filterDate('2024-06-01', '2024-08-31') \
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20)) \
    .median()

print("‚úÖ Sentinel-2 images fetched!")
print("   2020: Summer (Jun-Aug)")
print("   2024: Summer (Jun-Aug)")
```

```python
# Calculate NDVI for each period
ndvi_2020 = image_2020.normalizedDifference(['B8', 'B4'])
ndvi_2024 = image_2024.normalizedDifference(['B8', 'B4'])

# Calculate change
ndvi_change = ndvi_2024.subtract(ndvi_2020)

print("‚úÖ NDVI calculated!")
```

```python
# Mean NDVI for each period
stats_2020 = ndvi_2020.reduceRegion(
    reducer=ee.Reducer.mean(),
    geometry=area,
    scale=10
).getInfo()

stats_2024 = ndvi_2024.reduceRegion(
    reducer=ee.Reducer.mean(),
    geometry=area,
    scale=10
).getInfo()

mean_2020 = stats_2020['nd']
mean_2024 = stats_2024['nd']
change = mean_2024 - mean_2020

print(f"\nüìä STATISTICS:")
print(f"   Mean NDVI 2020: {mean_2020:.3f}")
print(f"   Mean NDVI 2024: {mean_2024:.3f}")
print(f"   Change: {change:+.3f} ({change*100:+.1f}%)")

if change > 0:
    print("   ‚úÖ Vegetation INCREASED")
else:
    print("   ‚ö†Ô∏è Vegetation DECREASED")
```

```python
# RGB (True Color)
rgb_params = {
    'bands': ['B4', 'B3', 'B2'],
    'min': 0,
    'max': 3000,
    'gamma': 1.4
}

# NDVI (Green gradient)
ndvi_params = {
    'min': 0,
    'max': 1,
    'palette': ['brown', 'yellow', 'lightgreen', 'darkgreen']
}

# Change (Red = loss, Green = gain)
change_params = {
    'min': -0.5,
    'max': 0.5,
    'palette': ['red', 'white', 'green']
}

print("‚úÖ Visualization parameters set!")
```

```python
print("Generating image URLs...")

# RGB images
url_rgb_2020 = image_2020.visualize(**rgb_params).getThumbURL({
    'region': area.getInfo(), 
    'dimensions': 512
})
url_rgb_2024 = image_2024.visualize(**rgb_params).getThumbURL({
    'region': area.getInfo(), 
    'dimensions': 512
})

# NDVI images
url_ndvi_2020 = ndvi_2020.visualize(**ndvi_params).getThumbURL({
    'region': area.getInfo(), 
    'dimensions': 512
})
url_ndvi_2024 = ndvi_2024.visualize(**ndvi_params).getThumbURL({
    'region': area.getInfo(), 
    'dimensions': 512
})

# Change map
url_change = ndvi_change.visualize(**change_params).getThumbURL({
    'region': area.getInfo(), 
    'dimensions': 512
})

print("‚úÖ URLs generated!")
```

```python
def download_image(url):
    """Download image from URL"""
    response = requests.get(url)
    return Image.open(BytesIO(response.content))

print("Downloading images...")

img_rgb_2020 = download_image(url_rgb_2020)
img_rgb_2024 = download_image(url_rgb_2024)
img_ndvi_2020 = download_image(url_ndvi_2020)
img_ndvi_2024 = download_image(url_ndvi_2024)
img_change = download_image(url_change)

print("‚úÖ Images downloaded!")
```

```python
# Create comparison plot
fig, axes = plt.subplots(2, 3, figsize=(16, 10))

# Row 1: RGB images
axes[0, 0].imshow(img_rgb_2020)
axes[0, 0].set_title('RGB - Summer 2020', fontweight='bold', fontsize=13)
axes[0, 0].axis('off')

axes[0, 1].imshow(img_rgb_2024)
axes[0, 1].set_title('RGB - Summer 2024', fontweight='bold', fontsize=13)
axes[0, 1].axis('off')

axes[0, 2].imshow(img_change)
axes[0, 2].set_title('NDVI Change\nüî¥ Loss | ‚ö™ No Change | üü¢ Gain', 
                     fontweight='bold', fontsize=13)
axes[0, 2].axis('off')

# Row 2: NDVI images
axes[1, 0].imshow(img_ndvi_2020)
axes[1, 0].set_title(f'NDVI - Summer 2020\n(Mean: {mean_2020:.3f})', 
                     fontweight='bold', fontsize=13)
axes[1, 0].axis('off')

axes[1, 1].imshow(img_ndvi_2024)
axes[1, 1].set_title(f'NDVI - Summer 2024\n(Mean: {mean_2024:.3f})', 
                     fontweight='bold', fontsize=13)
axes[1, 1].axis('off')

# Statistics panel
axes[1, 2].axis('off')
stats_text = f"""
üå≤ Poikilo Oros Forest
   Chaidari, Attica

üìä Statistics:

Summer 2020:
  NDVI = {mean_2020:.3f}

Summer 2024:
  NDVI = {mean_2024:.3f}

Change (2020‚Üí2024):
  {change:+.3f} ({change*100:+.1f}%)

{'‚úÖ Vegetation increased' if change > 0 else '‚ö†Ô∏è Vegetation decreased'}
"""

axes[1, 2].text(0.1, 0.5, stats_text, fontsize=12, 
                verticalalignment='center', family='monospace',
                bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

plt.suptitle('NDVI Comparison: Poikilo Oros Forest (2020 vs 2024)', 
             fontsize=16, fontweight='bold', y=0.98)
plt.tight_layout()
plt.show()
```
